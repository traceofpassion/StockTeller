```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/price/stock_price_kospi.csv',
            date_parser = 'Date',
            dtype = {
                'Open': np.int32,
                'High': np.int32,
                'Low': np.int32,
                'Close': np.int32,
                'Volume': np.int32,
                'Change': np.float64
                }
            )
code = pd.read_csv('../data/info/stock_info.csv')
```

```python
past = ('2019-12-01', '2019-12-31')
present = ('2020-11-20', '2020-12-04')
```

```python
df_past = data[(data['Date'] >= past[0]) & (data['Date'] <= past[1])]
df_present = data[(data['Date'] >= present[0]) & (data['Date'] <= present[1])]
```

```python
past_med = df_past.groupby('Symbol')['Close'].median()
present_med =df_present.groupby('Symbol')['Close'].median()
```

```python
df_total = past_med.to_frame('Past').merge(present_med.to_frame('Present'), how='left', on='Symbol')
```

```python
# Present/Past = Covdex (Smaller, Better)
df_total['Covdex'] = df_total['Present'] / df_total['Past']
```

```python
wish_price = (10000, 50000)
```

```python
df_wish = df_total[(df_total['Present'] >= wish_price[0]) & (df_total['Present'] <= wish_price[1])]
```

```python
n = 50
wishlist = df_wish.sort_values('Covdex').head(n)
idx = wishlist.index
```
